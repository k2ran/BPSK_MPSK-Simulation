```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
bimag=[]
img1=imageio.imread(r"C:\Users\Dhanush Joy\Desktop\cameraman.png")
plt.imshow(img1)
```

```python
fimag=[]
biarr=[]
print(img1)
print(img1.shape)
newim=img1.flatten()
print(max(newim))
for i in range(len(newim)):
    bimag.append(np.binary_repr(newim[i],width=8))
print(len(bimag))
```

```python
for i in range(len(bimag)):
    for j in bimag[i]:
        if j == '1':
            biarr.append(1)
        else:
            biarr.append(0)
print(biarr)
```

```python
for i in range(len(bimag)):
    for j in bimag[i]:
        if j == '1':
            fimag.append(1)
        else:
            fimag.append(-1)
len1=len(fimag)
print(len(fimag))
```

```python
plt.scatter(np.real(fimag),np.imag(fimag))
```

```python
snr_db=np.arange(-10,11,1)

def noise_ar(snr_db):
    sig=np.sqrt(1/(10**(snr_db/10)))
    noise=np.random.normal(0,sig/np.sqrt(2),len(fimag))+1j*np.random.normal(0,sig/np.sqrt(2),len(fimag))
    return noise

def Ber(biarr,reconstruct_arr):
    ber=0
    for i in range(len1):
        if biarr[i]!=reconstruct_arr[i]:
            ber+=1
    return ber/len(biarr)

def group(reconstruct_arr):
    pixel=[]
    len2=len(reconstruct_arr)
    for i in range(0, len2, 8):
        str1=reconstruct_arr[i:i+8]
        val = "".join(str(i) for i in str1)
        pixel.append(int(val,2))
    return pixel
```

```python
l=2
fig, ax = plt.subplots(21, 2, figsize=(15,40))
berar=[]
for i in range(len(snr_db)):
    noise_awgn=noise_ar(snr_db[i])
    rec_sig=fimag+noise_awgn
    
    
    plt.subplot(21,2,2*i+1)
    plt.scatter(np.real(rec_sig),np.imag(rec_sig))
    decode=[]
    for i in rec_sig:
        if np.real(i)>=0:
            decode.append(1)
        elif np.real(i)<0:
            decode.append(-1)
            
            
    reconstruct_arr=[]
    for i in decode:
        if i==-1:
            reconstruct_arr.append(0)
        else:
            reconstruct_arr.append(1)
            
            
    berval=Ber(biarr,reconstruct_arr)
    berar.append(berval)
    rec_Pixel=group(reconstruct_arr)
    rec_Pixel1=np.reshape(rec_Pixel,(256,256))
    rec_Pixel1=rec_Pixel1.astype(np.uint8)
    imageio.imwrite('hi.png',rec_Pixel1)
    image = imageio.imread('hi.png')
    plt.subplot(21,2,l)
    l=l+2
    plt.imshow(image)
```

```python
plt.plot(snr_db,berar)
```

```python
import math
berarrTh=[]
for i in snr_db:
    snr=10**(i/10)
    val=(1/2)*math.erfc(np.sqrt(snr))
    berarrTh.append(val)
plt.plot(snr_db,berarrTh)
```

```python
plt.semilogy(snr_db,berarrTh,'k',label='theoretical')
plt.semilogy(snr_db,berar,'k*',label='pratical')
plt.legend()
plt.show()
```
